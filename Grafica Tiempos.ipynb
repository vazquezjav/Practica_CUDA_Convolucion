```python
import numpy
import matplotlib
from matplotlib import pylab, mlab, pyplot
np = numpy
plt = pyplot
```

```python
x=["3x3","7x7","13x13"]
secuencial=[6.13,21.56,60.57]
paralelo=[0.121,0.233,0.397]
```

```python
plt.style.use("seaborn-whitegrid")
plt.plot(x,secuencial,'rd--')
plt.plot(x,paralelo,'bo--')
plt.xlabel('Imagen 920x730')
plt.ylabel('Tiempo milisegundos')
plt.legend(['CPU', 'CPU'], loc='upper left')
plt.show()
```
